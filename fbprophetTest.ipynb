```python
#Load the required libraries
import pandas as pd
import numpy as np
import seaborn as sns
from prophet import Prophet
```

```python
#Load the data
data = pd.read_csv('sorted_data_for_test.csv')
df = pd.DataFrame(data)
 
df = df.drop(['Recovered','Deaths','Province_State','Last_Update','Lat','Long_','Combined_Key','Incident_Rate','Case_Fatality_Ratio','Active','Year'],axis=1)

#View the data
df.head()
```

```python
df['FIPS'] = pd.to_datetime(df['FIPS'])
df.head()
```

```python
# Setting the Date values as index
df = df.set_index('FIPS')
```

```python
# to_datetime() method converts string
# format to a DateTime object
df.index= pd.to_datetime(df.index)
```

```python
start = df.index[0].date()

end = df.index[len(df)-1].date()

new_dates = pd.date_range(start=start,end=end,freq='D')
```

```python
df = df.reindex(new_dates)

df = df.rename_axis('FIPS')

df
```

```python
df = pd.DataFrame(df)
df.resample('D').interpolate() 
```
